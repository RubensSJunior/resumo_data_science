<h2>Amostra Sistemática</h2>
É uma amostra onde é escolhida um elemento aleatorio da população, apartir dele a cada N elementos um novo membro é escolhido.

```python
def amostraSistematica(data,jp,qtd):
    counter = 0
    choice = 0
    attributes = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]
    ndf = pd.DataFrame()
    while counter < qtd :
        choice = choice + jp
        try:
            ndf = ndf.append(data.iloc[choice],ignore_index=False)
            counter = counter+1
        except: 
            while choice > data['species'].count():
                choice = (choice-data['species'].count())
                if choice < 0:
                    choice = choice*-1
    return ndf
```

```python
#utilizando o mesmo 
amostraSistematica(dados,10,20)
```

<h1>Probabilidade</h1>
A probabilidade é denotada por P. O resultado do calculo de probabilidade varia entre 0 e 1,onde 0 é uma probabilidade impossivel e 1 um evento certo.Esse valor pode ser multiplicado por 100 pra encontrarmos a porcentagem da probabilidade.
<h2>Conceitos</h2>
<li><strong>Experimento:</strong> O que esta sendo estudado</li>
<li><strong>Espaço amostral:</strong> Todas as possibilidades de ocorrência do evento </li>
<li><strong>Evento:</strong> Resultados ocorridos</li>
<li><strong>Eventos Excludentes:</strong> Quando não podem ocorrer ao mesmo tempo Ex. Dado cair em um numero 3 ou par</li>
<li><strong>Eventos Não excludentes:</strong>Quando podem ocorrer ao mesmo tempo. Ex: dadi cair em m numero 3 ou impar</li>
<li><strong>Eventos Dependentes:</strong>A ocorrẽncia de um evento afeta outro. Um deles deve ocorrer para que outro ocorra.Ex: Possibilidade de jogar dois dados e dar 1 e 6</li>
<li><strong>Eventos Independentes:</strong>A ocorrência de um evento não afeta  outro Ex. Retirar um as e um 4 em uma mão com A,1,2,3,4,5,6</li>

A formula para a probabilidade é

P = Ocorrencia esperada/Numero de ocorrencias possiveis


<h2>Eventos Excludentes</h2>
Somas-se todas possibilidades
    1/6 (referente ao numero 3) + 3/6 (referente aos numeros pares) = 4/6 = 0.66

<h2>Eventos Não Exlcudentes</h2>
Somas-se todas possibilidades e subitraesse as sobreposições
    1/6 (referente ao numero 3) + 3/6 (referente aos numeros pares) - 1/6 (da sobreposição do numero 3 q é par) = 3/6 = 0.5

<h2>Eventos Independentes</h2>
    Aplicasse quando exisem mais de um eventos relacionados.
    Apartir dai se multiplica as possibilidades
    1/6 (posibilidade de tira 1 no dado) * 1/6 (possibilidade de tirar 6 no dado) = 1/36 => 0.027
<h2>Eventos Dependentes</h2>
    Aplicasse quando exisem mais de um eventos relacionados.
    Multiplica as possibilidades, subtraindo a amostragem referente a primeira escolha.
    1/6 (referente a primeira tirada) * 1/5 (referente a segunda escolha com uma carta a menos no espaço amostral) = 1/30 => 00.033

<h2>Probabilidade a "Longo Prazo"</h2>
Pequenas amostra tendem estar fora do padrão, entretanto com uma grande quantidade de casos para a media geralmente se mantem.<br>
Por exemplo a media do resultado de um lançamento de um dado é:
1+2+3+4+5+6/6 = 3,5

```python
import random
def testeProp(rng):
    x = 0.0
    for i in range(rng):
        x = x + random.randrange(1,7)
    x = x/rng
    return x
```

```python
#Podemos ver apartir da quantidade de interações o valor tende a se aproximar da media
testeProp(10)
```

<h2>Distribuição Binomial</h2>
É uma distribuição utilizada para inferir resultados dentro de dados discretos(Números fixos, inteiros e dentro de um intervalo.Ex: 1, 2, 3).
Pré requisitos:
<li>Numero fixo de experimentos</li>
<li>Numero cada experimeno por de ter dois resultados</li>
<li>A probabilidade deve ser a mesma</li>
<li>Eventos Independentes</li>

<h3>Convenções</h3>
<strong>x</strong> = total de sucessos<br>
<strong>p</strong> = probabilidade de sucesso<br>
<strong>n</strong> = número de experimentos<br>
<strong>1-p</strong> = Probabilidadde de fracassos<br>

A formula para essa expressão é:<br>
f(x) = (nx)p^x(1-p)^(n-x)<br><br>
Existe tambem uma tabela de distribuição binomial para auxiliar a busca.

```python
import math
#aplicando a formula temos a seguinte função
def distBin(x,n,p):
    return (math.factorial(n))/(math.factorial(x)*math.factorial(n-x))*(p**x)*((1-p)**(n-x))
```

```python
distBin(3,5,0.5)
```

```python
#Se somarmos todos os valores possiveis teremos o valor da probabilidade como 1
distBin(0,5,0.5)+distBin(1,5,0.5)+distBin(2,5,0.5)+distBin(3,5,0.5)+distBin(4,5,0.5)+distBin(5,5,0.5) 
```

<h1>Distribuição Normal</h1>
É uma distribuição onde a media esta no centro da amostra e 68.2% dos valores da amostra estão a até 1 desvio padrão da media, 17% esta entre 1 e 2 desvios padrão, 4.2% estão entre 2 e 3 desvios padrão e o rstante esta acima desse valor


<h2>Distribuição normal padrão </h2>
    <li>Mostra o numero de ddesvios padrões q o valor esta acima ou abaixo da media (Valor Z)
    <li>Média 0
    <li>Desvio Padrão 1
    <li> Usa-se uma fórmula para calcular a probabilidade de seus dados coom relação a tabela Z

Como o calculo de uma probabilidade utilizando uma distribuição normal pode ser complexo, utiliza-se a Distribuição normal padrão para simplificar o calculo.

A formula para a essa distribuição é:
Z = X-μ /  σ

```python
#Importando Biblioteca do Scipy para calculo do Z
import scipy.stats as st
#Função que efetua o calculo doa probabilidade
def desvioPadraoNormal(x,media,desvio,inverse=False):
    if inverse :
        return 100-(st.norm.cdf((x-media)/(desvio))*100)
    else :
        return st.norm.cdf((x-media)/(desvio))*100

#retorno da função com a probabilidade
desvioPadraoNormal(6,8,2)
```

```python
#Calculando o z apartir da probabilidade
round(st.norm.ppf(1-(1-0.95)/2),2)
```

```python
#Para calular na ordem inversa apenas insita o parametro 'inverse=True' no final
desvioPadraoNormal(10,8,2,inverse=True)
```

<h1>Intervalo de confiança</h1>
Alguns dos contras de se trabalhar com amostras é a possibilidade de suas inferências apresentarem erro e desvio padrão. Alem dos riscos da coleta de dados ruins e inviesamento.<br>
Entretando podemos medir o quanto podemos esperar essa variação.<br>

<li> intervalo de confiança: parametro + ou - a margem de erro estimada</li>
<li> Parametro: o valor a ser estimado (media/proporção)</li>
<li> Margem de erro: de 80 a 99%.É variabilidade possivel devido ao acaso</li>
<li> Tamanho da Amostra (n)</li>

<h2>Compensação</h2>
Aumentando o nivel de confiança o erro padrão tambem é aumenta e a margem de erro. Para dimnuir o erro padrão vc deve aumentar o numero da amostra.

```python
def intervaloConfianca(desvio,amostra,intConf):
    return round(round(st.norm.ppf(1-(1-(intConf/100))/2),2)*(desvio/(math.sqrt(amostra))),2)
```

```python
intervaloConfianca(1100,100,95)
```

```python
intervaloConfianca(1100,1000,95)
```

```python
intervaloConfianca(1100,100,99)
```

<h1>Troca entre margem de erro e, Intervalo de confiança e Amostra</h1>

Aumentando a margem de erro o meu intervalo de confiança é maior pois aumenta a chance da minha amostra estar dentro do intervalo
Da mesma forma aumentando o numero da amostra reduz a chance do efeito do acaso e a margem de erro reduz

```python
intervaloConfianca(1100,100,95)
```

<h1>Teste de Hipótese</h1>
Serve para confirmar uma premisa de uma amostra. Como uma amostra é passivel de erros é necessario testar os resultados tem uma diferença é estatisticamente significativa.

<li>H0 = Hipótese nula: A alegação que se quer testar</li>
Presumesse sempre que H0 é verdadeiro, a não ser que existão evidencias para provar o contrario
<li>Ha = hipótese alternativa</li>

<h2>Teste de Hipotese</h2>
Score Padrão: Os erros que seus dados apresentão que estão abaixo ou acima da média. è calculado como a Media dividido pelo erro padrão

<h2>Níveis de Alfa e valor-p</h2>
Alfa é um valor utilizado para o seu teste de hipótese.
Se o valor-p é MAIOR OU IGUAL ao alfa voce nao deve rejeitar o H0(Hipótee nula)
Se o valor-p for MENOR que o alfa voce deve rejeitar H0
Se o valor for bem proximo é dito que o teste é inconclusivo

<h2>Etapas</h2>
<ol>
    <li>Definir o tamanho da amostra</li>
    <li>Coletar os dados</li>
    <li>Calcular a média e o Desvio padrão</li>
    <li>Definir duas hípoteses</li>
    <li>Definir o alfa</li>
    <li>Padronizar os dados gerando a estatistica de teste</li>
    <li>Encontrar valor-p na Tabela Z</li>
    <li>comparar com o valor do alfa</li>
    <li>Emitir veredito</li>
</ol>

```python
def valorPMedia(amostra,desvio,media,hipotese,maior=False):
    vp = (media-hipotese)/(desvio/(math.sqrt(amostra)))
    zs = round(st.norm.cdf(vp),4)
    if maior :
        return  round(1-zs,4)
    else :
        return  round(zs,4)
```

```python
valorPMedia(100,4,23,22,maior=True)
```

<h2>Tipos de Erros</h2>
Em um teste de hipoteses é possivel encontrar dois erros:
<li>Erro do tipo 1: Rejeitar H0 quando não deveria</li>
<li>Erro do tipo 2: Não rejeitar H0 quando deveria ter rejeitado</li>
<h2>Trocas</h2>
Diminuindo o valor do alfa, voce reduz a chance do erro do tipo 1 pois torna mais dificil a rejeição do H0, porem aumenta a chance do erro do tipo 2. Aumentando o valor de Alfa a possibilidade do erro do tipo 2 tambem diminui, pois torna mais facil o rejeitar o h0 , concequentemente aumenta a chance do erro do tipo um. 
Essas trocas dependem da necessidade dentro da hipoteses que voce quer testar. Por exemplo uma empresa farmaceutica não pode deixar um erro do tipo 2 acontecer no teste de remedios por exemplo

<h1>T de Student</h1>
É um tipo de distribuição utilizada para amosta pequenas (menores que 30) e tambem quando não se conhece o desvio padão da população. Possui uma maior variabilidade alem de possui caudas maiores tambem.<br>
T de Student utiliza uma um conceito chamado de Grau de liberdade que é a amostra menos 1.<br>
Com essa distribuição é possuvel calcular probabilidades, intervalos de confiança e teste de hipoteses porem só é utilizavel com amostra menores que 30. Com ela podemos:
<li>Calculamos o valor T</li>
<li>Consultamos a tabela de distrinuição T</li>




```python
def tStudentProbabilidade(media,amostra,desvio,p,maior=False):   
    t = (p-media)/(desvio/(math.sqrt(amostra)))
    if maior:
        return 1-t
    else :
        return t
```

```python
tStudentProbabilidade(75,9,10,80)
```
