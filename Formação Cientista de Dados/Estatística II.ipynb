```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

<h1>Introdução a Correlação e Regressão Linear</h1>
Atravez da analise de duas ou mais variáriaveis é possivel descobrir se existe algum tipo de relação matematica entre elas alem de descobrir a força dessa relação. Com essa relação é possivel fazer uma previsão de valores dentro do espectro das variáveis analisadas. Atravez da Regressão Linear supõe que existe uma linearidade nos dados. Atravez dessa linearidade é possivel criar uma linha de regressão para prever os resultados.

<h2>Gráfico de Disperção</h2>
É um grafico que auxilia a visualização par a par dos dados.

```python
#Criação dos campos a serem relacionados
idade = pd.Series([18,23,28,33,38,43,48,53,58,63])
custo = pd.Series([871,1132,1242,1356,1488,1638,2130,2454,3066,4090])
#Criação do DataFrame com as analises
data = pd.DataFrame({'Idade' : idade,'Custo' :custo })
#Exibindo os Gráficos
plt.scatter(data['Idade'],data['Custo'])
plt.show()
```

<h1>Correlação (R)</h1>

A correlação mostra a força de uma relação entre os dados, ela é representada por R.
Essa correlação demonstra não só a força dessa correlação como tambem a direção dessa correlação.A correlação pode ser proporcional (ambbos os numeros seguem a mesma direção ) ou inversa (os numeros seguem em direções opostas). essa correlação deve estar entre os valores 1 e -1. Quanto mais proximo a  menor a força da correlação.

<li>1 : Perfeita</li>
<li>0.7 : Forte</li>
<li>0.5  : Moderada</li>
<li>0.25 : Fraca</li>
<li>0 : Inexistente</li>
<li>-0.25 : Fraca</li>
<li>-0.5  : Moderada</li>
<li>-0.7 : Forte</li>
<li>-1 : Perfeita</li>


<h2>Coeficiente de determinação (R²)</h2> 
Ele representa o quanto um modelo pode explicar os valores, quanto maior, mais explicativo ele é. O restante da variabilidade é determinado por demais variáveis que nao estão no modelo descrito. 
Ela sempre Varia entre 0 e 1 e é calculado com o quadrado do coeficiente de correlação.
