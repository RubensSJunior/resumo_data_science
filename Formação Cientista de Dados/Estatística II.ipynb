<h1>Introdução a Correlação e Regressão Linear</h1>
Atravez da analise de duas ou mais variáriaveis é possivel descobrir se existe algum tipo de relação matematica entre elas alem de descobrir a força dessa relação. Com essa relação é possivel fazer uma previsão de valores dentro do espectro das variáveis analisadas. Atravez da Regressão Linear supõe que existe uma linearidade nos dados. Atravez dessa linearidade é possivel criar uma linha de regressão para prever os resultados.

<h2>Gráfico de Disperção</h2>
É um grafico que auxilia a visualização par a par dos dados.
<h1>Correlação (R)</h1>

A correlação mostra a força de uma relação entre os dados, ela é representada por R.
Essa correlação demonstra não só a força dessa correlação como tambem a direção dessa correlação.A correlação pode ser proporcional (ambbos os numeros seguem a mesma direção ) ou inversa (os numeros seguem em direções opostas). essa correlação deve estar entre os valores 1 e -1. Quanto mais proximo a  menor a força da correlação.

<li>1 : Perfeita</li>
<li>0.7 : Forte</li>
<li>0.5  : Moderada</li>
<li>0.25 : Fraca</li>
<li>0 : Inexistente</li>
<li>-0.25 : Fraca</li>
<li>-0.5  : Moderada</li>
<li>-0.7 : Forte</li>
<li>-1 : Perfeita</li>


<h2>Coeficiente de determinação (R²)</h2> 
Ele representa o quanto um modelo pode explicar os valores, quanto maior, mais explicativo ele é. O restante da variabilidade é determinado por demais variáveis que nao estão no modelo descrito. 
Ela sempre Varia entre 0 e 1 e é calculado com o quadrado do coeficiente de correlação.
<h2>Valor Residual</h2>
É a distancia entre o ponto apresentado no grafico e a linha de regressão trassada no grafico.
Se o ponto estiver acima da linha o valor do residuo é positivo, se estiver abaixo é negativo. o valor do ponto mais o valor do residuo é chamado de VALOR AJUSTADO
</h2>Outlaiers</h2>
São valores anormais que fojem do padrão do gráfico, eles podem influenciar na correlação e alterar a precisão de qualquer modelo criado. Apartir da analise de cada cenario é possivel definir qual a melhor ação a ser tomada para utilização do modelo.
<h2>Extrapolação</h2>
É o trabalho com os dados alem dos valores apresentados. Esses valores devem ser analisados para ver se a analise faz sentido ou se a regra de linearidade aplica a eles.
<h2>Correlaçã Não é Causa</h2>
É o entendimento de que não é por que duas variaveis possuem correlações matemáticas que elas causam uma a outra, elas podem não necessariamente representar uma relação direta entre os numeros.
<h2>Condições</h2>
Para fazer predissões utilizando regressão linear é necessario atender alguns criterios como:
<li>Existir uma correlação forte ou moderada independente do sinal</li>
<li>Coeficiente de determinação qualquer valor acima de 7 é considerado otimo, valores menores que 3 devem ser descartados e  valores entre essa faixa é necessario analisar o cenario para entender o caso</li>
<li>Residuais Padronizados> Residuais devem ter estar proximos a uma distribuição normal</li>
